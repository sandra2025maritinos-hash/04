```
import time
import sys

# função para o efeito de máquina de escrever (mais rápido e discreto agora)
def imprimir_efeito(texto, atraso=0.015):
    for letra in texto:
        sys.stdout.write(letra)
        sys.stdout.flush()
        time.sleep(atraso)
    print()

def cabecalho():
    print("-" * 50)
    imprimir_efeito("aula: como nasce uma ia (versão compacta)", 0.02)
    print("-" * 50)
    time.sleep(1)

def passo_1():
    print("\n[ passo 1: a dieta da ia - coleta de dados ]")
    print("código fonte:")
    codigo = """def nascer_e_aprender():
    textos_do_mundo = baixar_internet_inteira()
    return textos_do_mundo"""
    print(codigo)

    print("\nexplicação: a variável 'textos_do_mundo' guarda petabytes de livros e sites.")
    print(">> a avançar para a execução em 3s...\n")
    time.sleep(3)

    imprimir_efeito("run: a descarregar a wikipedia e artigos científicos... [ok]")
    time.sleep(1)

def passo_2():
    print("\n[ passo 2: mastigar palavras - tokenização ]")
    print("código fonte:")
    codigo = """frase = "eu sou uma ia"
tokens = transformar_em_numeros(frase)
print(tokens)"""
    print(codigo)

    print("\nexplicação: computadores não leem letras. a string vira números (tokens).")
    print(">> a avançar para a execução em 3s...\n")
    time.sleep(3)

    imprimir_efeito("run: a converter 'ia' -> token [8821]... [ok]")
    time.sleep(1)

def passo_3():
    print("\n[ passo 3: o cérebro - arquitetura transformer ]")
    print("código fonte:")
    codigo = """class Cerebro:
    def __init__(self):
        self.conexoes = 150000000000  # 150 mil milhões
        self.atencao = focar_no_contexto()"""
    print(codigo)

    print("\nexplicação: criamos o objeto principal com milhares de milhões de parâmetros.")
    print(">> a avançar para a execução em 3s...\n")
    time.sleep(3)

    imprimir_efeito("run: a montar a rede neural e o mecanismo de atenção... [ok]")
    time.sleep(1)

def passo_4():
    print("\n[ passo 4: a escola - treino e backpropagation ]")
    print("código fonte:")
    codigo = """for tentativa in range(1000000):
    chute = adivinhar_palavra("o céu é...")
    if chute != "azul":
        aprender_com_erro()"""
    print(codigo)

    print("\nexplicação: um laço de repetição (for) treina a ia a adivinhar palavras.")
    print(">> a avançar para a execução em 3s...\n")
    time.sleep(3)

    imprimir_efeito("run: tentativa 1 (erro) -> a ajustar matemática da rede... [ok]")
    imprimir_efeito("run: tentativa 5000 (acerto) -> conhecimento consolidado... [ok]")
    time.sleep(1)

def passo_5():
    print("\n[ passo 5: bons modos - alinhamento rlhf ]")
    print("código fonte:")
    codigo = """def verificar_seguranca(resposta):
    if resposta == "tóxica":
        punir_ia()
    elif resposta == "útil":
        recompensar_ia()"""
    print(codigo)

    print("\nexplicação: humanos avaliam as respostas para garantir que a ia é segura e educada.")
    print(">> a avançar para a execução em 3s...\n")
    time.sleep(3)

    imprimir_efeito("run: a aplicar feedback humano... personalidade ajustada... [ok]")
    time.sleep(1)

def conclusao():
    print("\n" + "-" * 50)
    imprimir_efeito("processo concluído. a ia está pronta a ajudar!")
    print("-" * 50)

if __name__ == "__main__":
    cabecalho()
    passo_1()
    passo_2()
    passo_3()
    passo_4()
    passo_5()
    conclusao()
```
