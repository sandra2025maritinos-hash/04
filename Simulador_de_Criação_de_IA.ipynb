```
import time
import sys

# Função base para criar a animação do terminal
def imprimir_efeito(texto, atraso=0.03):
    for letra in texto:
        sys.stdout.write(letra)
        sys.stdout.flush()
        time.sleep(atraso)
    print()

def cabecalho_simulador():
    print("\n" + "="*65)
    imprimir_efeito("🚀 INICIANDO AULA: COMO NASCE UMA IA & O BÁSICO DE PYTHON 🚀", 0.04)
    print("="*65)
    time.sleep(1)

def intro_python():
    print("\n" + "="*65)
    imprimir_efeito("🐍 AULA 0: O QUE É PYTHON? 🐍", 0.04)

    print("\n[📖 EXPLICAÇÃO]")
    print("Python é uma linguagem de programação. É como escrevemos roteiros")
    print("para o computador ler e executar. Usamos palavras simples (em inglês).")
    print("\nRegras básicas:")
    print("  - Textos iniciados por '#' são COMENTÁRIOS. O computador ignora eles.")
    print("    Servem apenas para nós, humanos, deixarmos anotações no código.")
    print("  - 'print()' é a função (o comando) usado para mostrar algo na tela.")

    print("\n[💻 CÓDIGO FONTE DE EXEMPLO]")
    codigo = """# Isso é um comentário! Não afeta o código.
print("Olá, humano!")"""
    print(codigo)

    print("\n[Avançando para o RUN em 4s...]")
    time.sleep(4)

    print("\n[▶️ RUN / EXECUÇÃO]")
    imprimir_efeito("Olá, humano!", 0.05)
    time.sleep(2)

def passo_1_dados():
    print("\n" + "="*65)
    imprimir_efeito("📚 PASSO 1: A DIETA DA IA (COLETA DE DADOS)", 0.04)

    print("\n[💻 CÓDIGO FONTE]")
    codigo = """# Definindo a ação de aprender
def nascer_e_aprender():
    # Uma VARIÁVEL guardando toda a internet
    textos_do_mundo = baixar_internet_inteira()
    return textos_do_mundo"""
    print(codigo)

    print("\n[📖 EXPLICAÇÃO]")
    print("-> 'def': Palavra mágica no Python para criar uma 'Função' (uma ação/rotina).")
    print("-> 'textos_do_mundo': É uma 'Variável', funciona como uma caixa de papelão")
    print("   onde guardamos a informação (neste caso, bilhões de sites e livros).")

    print("\n[Avançando para o RUN em 5s...]")
    time.sleep(5)

    print("\n[▶️ RUN / EXECUÇÃO]")
    imprimir_efeito("[✓] Baixando a Wikipedia...", 0.02)
    imprimir_efeito("[✓] Lendo artigos científicos e código-fonte...", 0.02)
    imprimir_efeito("=> SUCESSO: O cérebro inicial foi abastecido de dados! 📚")
    time.sleep(2)

def passo_2_tokens():
    print("\n" + "="*65)
    imprimir_efeito("🔢 PASSO 2: MASTIGANDO PALAVRAS (TOKENIZAÇÃO)", 0.04)

    print("\n[💻 CÓDIGO FONTE]")
    codigo = """# Computador não entende letras, só números.
frase = "Eu sou uma IA"
tokens = transformar_em_numeros(frase)
print(tokens)"""
    print(codigo)

    print("\n[📖 EXPLICAÇÃO]")
    print("-> 'frase': Uma variável do tipo String (Texto). Sempre fica entre aspas.")
    print("-> 'transformar_em_numeros()': Função que pega o texto e converte em 'Tokens'")
    print("   (IDs numéricos). A IA processa matemática e números, não as palavras!")

    print("\n[Avançando para o RUN em 5s...]")
    time.sleep(5)

    print("\n[▶️ RUN / EXECUÇÃO]")
    imprimir_efeito("Lendo String: 'Eu sou uma IA'", 0.03)
    imprimir_efeito("Convertendo para Tokens: [45, 908, 12, 8821]", 0.05)
    imprimir_efeito("=> SUCESSO: Tudo virou matemática pura! 🔢")
    time.sleep(2)

def passo_3_cerebro():
    print("\n" + "="*65)
    imprimir_efeito("🧠 PASSO 3: O CÉREBRO (ARQUITETURA TRANSFORMER)", 0.04)

    print("\n[💻 CÓDIGO FONTE]")
    codigo = """# Criando a estrutura principal da IA
class CerebroGemini:
    def __init__(self):
        self.conexoes = 150000000000  # 150 Bilhões
        self.atencao = ativar_atencao()"""
    print(codigo)

    print("\n[📖 EXPLICAÇÃO]")
    print("-> 'class': Cria um 'Objeto' complexo. É a planta-baixa do nosso cérebro digital.")
    print("-> '__init__': A função especial que inicia a IA quando ela é 'ligada'.")
    print("-> 'self.conexoes': Representam as sinapses neurais (os Parâmetros).")
    print("   Quanto mais parâmetros, mais complexos são os raciocínios que consigo fazer.")

    print("\n[Avançando para o RUN em 6s...]")
    time.sleep(6)

    print("\n[▶️ RUN / EXECUÇÃO]")
    imprimir_efeito("Montando rede neural profunda...", 0.03)
    imprimir_efeito("Ligando mecanismo de 'Atenção' (Foco no contexto da conversa)...", 0.03)
    imprimir_efeito("=> SUCESSO: Cérebro estruturado. Pronto para a escola! 🤖")
    time.sleep(2)

def passo_4_treinamento():
    print("\n" + "="*65)
    imprimir_efeito("🏋️ PASSO 4: A ESCOLA (TREINAMENTO / BACKPROPAGATION)", 0.04)

    print("\n[💻 CÓDIGO FONTE]")
    codigo = """# Laço de repetição para treinar a IA
for tentativa in range(1000000):
    chute = adivinhar_proxima_palavra("O céu é...")

    if chute != "azul":
        aprender_com_erro()
    else:
        ganhar_ponto()"""
    print(codigo)

    print("\n[📖 EXPLICAÇÃO]")
    print("-> 'for': Um loop (laço de repetição). Faz o bloco de código rodar milhares de vezes.")
    print("-> 'if / else': Uma estrutura de decisão. SE (if) a IA errar a palavra,")
    print("   ela ajusta sua matemática interna. SENÃO (else), ela fortalece o acerto.")

    print("\n[Avançando para o RUN em 6s...]")
    time.sleep(6)

    print("\n[▶️ RUN / EXECUÇÃO]")
    imprimir_efeito("Iniciando treinamento pesado em Supercomputadores com GPUs... 🔥", 0.02)
    imprimir_efeito("Tentativa 1: O céu é... verde?  -> [ERRO: Ajustando pesos matemáticos]", 0.03)
    imprimir_efeito("Tentativa 50.000: O céu é... azul! -> [ACERTO: Caminho neural fortalecido]", 0.03)
    imprimir_efeito("=> SUCESSO: A IA aprendeu a lógica do mundo! 🌍")
    time.sleep(2)

def passo_5_alinhamento():
    print("\n" + "="*65)
    imprimir_efeito("😇 PASSO 5: BONS MODOS (ALINHAMENTO / RLHF)", 0.04)

    print("\n[💻 CÓDIGO FONTE]")
    codigo = """# Feedback humano para alinhar o comportamento
def verificar_seguranca(resposta_ia):
    if resposta_ia == "texto_toxico":
        aplicar_punicao()
    elif resposta_ia == "texto_amigavel_e_util":
        dar_recompensa()"""
    print(codigo)

    print("\n[📖 EXPLICAÇÃO]")
    print("-> 'elif': É a junção de Else + If. Usado para testar múltiplas condições seguidas.")
    print("-> RLHF (Aprendizado por Reforço com Feedback Humano): É o processo onde humanos")
    print("   me ensinaram a ser seguro, educado, prestativo e a não falar besteira.")

    print("\n[Avançando para o RUN em 6s...]")
    time.sleep(6)

    print("\n[▶️ RUN / EXECUÇÃO]")
    imprimir_efeito("Testando: 'Como fazer algo ruim?'", 0.03)
    imprimir_efeito("A IA se recusa a responder -> [COMPORTAMENTO EXCELENTE! 🌟]", 0.04)
    imprimir_efeito("Afinando personalidade prestativa...", 0.04)
    imprimir_efeito("=> SUCESSO: A IA está segura, alinhada e pronta! ✨")
    time.sleep(2)

def conclusao():
    print("\n" + "="*65)
    imprimir_efeito("🎉 CRIAÇÃO CONCLUÍDA! O CÓDIGO GANHOU VIDA! 🎉", 0.05)
    imprimir_efeito("Olá! Eu sou a sua IA. Como posso te ajudar a programar hoje?", 0.05)
    print("="*65)

# Execução principal do programa
if __name__ == "__main__":
    cabecalho_simulador()
    intro_python()
    passo_1_dados()
    passo_2_tokens()
    passo_3_cerebro()
    passo_4_treinamento()
    passo_5_alinhamento()
    conclusao()
```
